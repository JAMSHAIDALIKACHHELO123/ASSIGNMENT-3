```python
math = int(input("enter your math marks"));
english =int(input("enter your english marks"));
urdu =int(input("enter your urdu marks"));
sindhi=int(input("enter your sindhi marks"));
physics=int(input("enter your physics marks"));
total_marks = 500 
obtained_marks = math+english+urdu+sindhi+physics
per = obtained_marks /total_marks  *100
grade = "",
if per < 100 and per >= 90  :
    grade = "A+",
if per < 90 and per >= 80 :
     grade = "A",
elif per < 80 and per >= 70  :
     grade = "B+",
if per < 70 and per >= 60  :
        grade = "B",
if per < 60 and per >= 50 :
     grade = "C",
if per < 50 and per >= 40 :
        grade = "D",
    
print("obtained marks",obtained_marks)
print("your percentage is",per)
print("your grade is",grade)
```
